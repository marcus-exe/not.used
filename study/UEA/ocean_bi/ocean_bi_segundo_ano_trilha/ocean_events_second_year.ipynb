```python
from utils import *
from utils.slugify import slugify

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker

import pandas as pd
import re 

%matplotlib inline
```

```python
def search_word(df,search_column,palavra):
    new_df = df.loc[df[search_column].str.contains(palavra,flags=re.IGNORECASE)]
    return new_df
```

```python
df1 = pd.read_csv("ocean_online_events_second_year.csv")
df2 = pd.read_csv("cursos_ofertados_1_ano_com_trilha.csv")
```

```python
#limpando os dados de título de curso 

#removendo espaços nos nomes do curso
new_values = df1["eve_title"].apply(lambda data: re.sub(r"\s","",data).lower())
df1["course_name"] = new_values
```

```python
##criando a coluna de trilhas 
df1["eve_trilha"] = 0
df1.columns
```

```python
df1.course_name
```

```python
#removendo cursos não interessantes 
excluded_courses = ["ux","email","indústria4.0","cidadesinteligentes","usabilidade","designthinking"]
for word in excluded_courses:
    excluded = search_word(df1,"course_name",word)
    df1.drop(excluded.index,axis=0,inplace=True)
```

```python
df1.eve_title.unique()
```

```python
#excluding more irrevalant courses
more_excluded_courses = ["HealthTechs","Mindset"]
for word in more_excluded_courses:
    excluded = search_word(df1,"course_name",word)
    df1.drop(excluded.index,axis=0,inplace=True)
```

```python
df1.eve_title.unique()
```

```python
#definindo as trilhas
trilhas = {"IAR":"IAR","BXB":"BXB","IOT":"IOT","TZN":"TZN","BLC":"BLC",
           "GME":"GME","LAB":"LAB","NGC":"NGC","RVA":"RVA","TALK":"TALK",
           "PRG":"PRG","AND":"AND","ÁGIL":"ÁGIL","INT":"INT","FRTEND":"FRTEND","WRB":"WRB","BCKEND":"BCKEND","FABLAB":"FABLAB"}
```

```python
def set_trilha(df,df_search,trilha):
    df_search.loc[:,"eve_trilha"]= trilha
    df.loc[df_search.index,"eve_trilha"]=trilha
    return df
```

```python
#função para definir sessões 
def define_session(data):
    session = re.findall(r"\d",data)
    if len(session)>0:
        return session[0]
    else:
        return "-"
```

```python
#criando a coluna is_session para definir os cursos com mais de uma sessão
df1["is_session"] = 0
new_values = df1["eve_title"].apply(lambda data: define_session(data))
df1["is_session"] = new_values
```
